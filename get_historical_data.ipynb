```python
import yfinance as yf
import pandas as pd
from IPython.display import display
```

## Get the data from yahoo

```python
#choose the security 
SSO = yf.Ticker("SSO")

#ask for the historical data
SSO_hist = SSO.history(period="max")
#those columns are all 0
SSO_hist = SSO_hist.drop(['Dividends', 'Stock Splits'], axis = 1)
pd.set_option('display.max_columns', None)

#display the dataframe
display(SSO_hist)
```

```python
# create a function to add all the value to a dataframe
def create_dataframe(list_securities):
    ''' input: list_securities = list with teh name of the security as string
        output: pandas dataframe with column=security, value=close_price, index=date(y%m%d) 
        '''
    df = pd.DataFrame()
    for security in list_securities:
        SEC = yf.Ticker(security)
        hist = SEC.history(period="max")
        df[security] = hist['Close'] #I just consider the close price, is that correct?
        #df.dropna() #if this line the data are only from 2010 otyherwise some columns arrive until 2006
    return df
```

```python
# try of the function
df = create_dataframe(['SSO','UBT', 'UST', 'UGL', 'DIG'])
df
```
